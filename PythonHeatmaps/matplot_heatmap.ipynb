```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

attributes = ['Women Entrepreneurship Index', 'Entrepreneurship Index', 'Inflation rete', 'Female Labor Force Participation Rate']
df = pd.DataFrame([
    [1.0, 0.91, -0.46, 0.44],
    [0.91, 1, -0.4, 0.33],
    [-0.46, -0.4, 1, -0.14],
    [0.44, 0.33, -0.14, 1],
], columns=attributes, index=attributes)

fig, ax = plt.subplots()
im = ax.imshow(df)

ax.set_xticks(np.arange(len(df.columns)))
ax.set_xticklabels(df.columns)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

ax.set_yticks(np.arange(len(df.index)))
ax.set_yticklabels(df.index)

for i in range(len(df.columns)):
    for j in range(len(df.index)):
        text = ax.text(j, i, df.iloc[i, j], ha="center", va="center", color="w")

ax.set_title("Corr")
plt.show()
```
